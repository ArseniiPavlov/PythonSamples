```python
import numpy as np
import networkx as nx
from matplotlib import pyplot as plt
%pylab inline
```

```python
G = nx.dense_gnm_random_graph(45,95)
```

```python
len(G.edges)
```

```python
nx.draw_networkx(G)
```

```python
visited_nodes = []
node_start = 0
node_end = 11
runs = 1000
ways = []

for run in xrange(runs):
    node = node_start
    step = 0
    ways.append([0])
    while node!=node_end:
        node = G[node].keys()[np.random.randint(0, len(G[node].keys()))]
        ways[run].append(node)
        step += 1
    visited_nodes.append(step)
```
